```python
import matplotlib.pyplot as plt
from pylab import mpl 
from dateutil.parser import parse
mpl.rcParams['font.sans-serif'] = ['SimHei']
# 以下代码从全局设置字体为Arial Unicode MS，解决显示中文问题【mac】
# 设置font.sans-serif 或 font.family 均可
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']
# plt.rcParams['font.family']=['Arial Unicode MS']
# 解决中文字体下坐标轴负数的负号显示问题
plt.rcParams['axes.unicode_minus'] = False 

Keyword="ActionGeo_ADM1Code"
```

```python
df['Day'] = df['Day'].astype(str)
```

```python
date_standard=0
for i in df.Day:
    df.Day[date_standard]=parse(str(i))
    date_standard=date_standard+1
```

```python
df.Day
```

```python
#按年
#df['Day'] = df['Day'].astype(str)

date=df[["Day",Keyword]]
date.dropna(axis=0, how='any', inplace=True)
date=date["Day"].value_counts()
date=date.sort_index(axis=0, ascending=True)
date.to_csv("DateCount1.csv",header=1)
```

```python
date[32:]
```

```python
data1=date[32:]
```

```python

import numpy as np
plt.figure(figsize=(20, 8), dpi=120)
# 归一化
normalized_data = (date1 - date1.min()) / (date1.max() - date1.min())
plt.plot(normalized_data,label='Original Curve',color='#F5F5F5')

#平滑曲线
from scipy.interpolate import make_interp_spline 
from scipy.signal import savgol_filter
tmp_smooth1 = savgol_filter(normalized_data.values,80,5, mode= 'nearest')

plt.plot(normalized_data.index,tmp_smooth1,label='Filtered Smooth Curve',color='#771693',linewidth=1.5)

#制图
plt.xlabel('Day')
plt.ylabel('Normalized Count')
plt.title('Trends in Decolonization in Health Policy')
plt.legend()
plt.savefig('World.svg')
plt.show()
```
